```python
import pandas as pd
```

```python
#First load the data from a csv dataset
nbaData = pd.read_csv("nba_games.csv", index_col=0)
```

```python
nbaData
```

```python
#Sort by data
nbaData = nbaData.sort_values('date')
```

```python
#Reset index
nbaData = nbaData.reset_index(drop=True)
```

```python
nbaData
```

```python
#deleting not important rows
del nbaData["mp.1"]
del nbaData["mp_opp.1"]
del nbaData["index_opp"]
nbaData
```

```python
#over here, we create a new column called target, which gets the next games outcome
def add_target(team):
    team["nextGame"] = team["won"].shift(-1)
    return team

#Target is the next games match outcome, so we shift the next games match result into its own column
nbaData = nbaData.groupby("team", group_keys=False).apply(add_target)
```
